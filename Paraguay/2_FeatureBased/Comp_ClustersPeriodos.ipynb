# COMPARACIÓN porcentual entre cluster Y PUNTAJES POR PERIODO
### FeatureBased(T) vs FeatureBased(P1)
### FeatureBased(T) vs FeatureBased(P2)

```python
import json
import pandas as pd
import os

my_path = os.path.abspath('')
my_path = my_path.split('\\')
my_path_py = "\\".join(my_path[:-1])

grupo_clusterFB = pd.read_csv("ClusterFB1_2009al2013.csv")
#grupo_clusterFB = grupo_cluster.drop('Unnamed: 0', axis=1)
grupo_clusterRB = pd.read_csv(my_path_py+"\\1_RawBased\\ClusterRB_2009al2013.csv")
grupo_clusterRB = grupo_clusterRB.drop('Unnamed: 0', axis=1)
```

```python
print(len(grupo_clusterRB))
print(len(grupo_clusterFB))
```

```python
df_comp = pd.DataFrame()
df_comp['Distrito'] = grupo_clusterFB['Distrito']
df_comp['KM RB'] = grupo_clusterRB['Cluster KM']
df_comp['KM FB'] = grupo_clusterFB['Cluster KM']
df_comp['HAC RB'] = grupo_clusterRB['Cluster HAC ']
df_comp['HAC FB'] = grupo_clusterFB['Cluster HAC ']
df_comp['DB RB'] = grupo_clusterRB['Cluster DB SP']
df_comp['DB FB'] = grupo_clusterFB['Cluster DB SP']
df_comp
```

# Comparación RB y FB

```python
df_comp['Match']= df_comp['KM RB'] == df_comp['KM FB']
print("Variación de valores del RawBased al Feature Based en K-Means: ")
print(df_comp['Match'].value_counts()[False]/200*100, "%")

df_comp['Match']= df_comp['HAC RB'] == df_comp['HAC FB']
print("Variación de valores del RawBased al FeatureBased en Jerárquico Aglomerativo: ")
print(df_comp['Match'].value_counts()[False]/200*100, "%")

df_comp['Match']= df_comp['DB RB'] == df_comp['DB FB']
print("Variación de valores del RawBased al FeatureBased en DBSCAN: ")
print(df_comp['Match'].value_counts()[False]/200*100, "%")
```

# Comparación por períodos FB

```python
grupo_clusterFB1 = pd.read_csv("ClusterFB1_2009-2010-2011.csv")
grupo_clusterFB1 = grupo_clusterFB1.drop('Unnamed: 0', axis=1)
grupo_clusterFB2 = pd.read_csv("ClusterFB1_2012-2013.csv")
grupo_clusterFB2 = grupo_clusterFB2.drop('Unnamed: 0', axis=1)
print(len(grupo_clusterFB1))
print(len(grupo_clusterFB2))
```

```python
#grupo_clusterFB1[['Distrito','Cluster KM']]
grupo_clusterFB1 = grupo_clusterFB1.rename(columns={"Cluster KM": "KM P1"})
grupo_clusterFB1 = grupo_clusterFB1.rename(columns={"Cluster HAC ": "HAC P1"})
grupo_clusterFB1 = grupo_clusterFB1.rename(columns={"Cluster DB SP": "DB P1"})
grupo_clusterFB2 = grupo_clusterFB2.rename(columns={"Cluster KM": "KM P2"})
grupo_clusterFB2 = grupo_clusterFB2.rename(columns={"Cluster HAC ": "HAC P2"})
grupo_clusterFB2 = grupo_clusterFB2.rename(columns={"Cluster DB SP": "DB P2"})
grupo_clusterFB1
```

```python
df_comp = pd.DataFrame()
df_comp['Distrito'] = grupo_clusterFB['Distrito']
df_comp['KM Total'] = grupo_clusterFB['Cluster KM']
df_comp = pd.merge(df_comp, grupo_clusterFB1[['Distrito','KM P1']], how="outer", on=["Distrito"])
df_comp = pd.merge(df_comp, grupo_clusterFB2[['Distrito','KM P2']], how="outer", on=["Distrito"])
df_comp['HAC Total'] = grupo_clusterFB['Cluster HAC ']
df_comp = pd.merge(df_comp, grupo_clusterFB1[['Distrito','HAC P1']], how="outer", on=["Distrito"])
df_comp = pd.merge(df_comp, grupo_clusterFB2[['Distrito','HAC P2']], how="outer", on=["Distrito"])
df_comp['DB Total'] = grupo_clusterFB['Cluster DB SP']
df_comp = pd.merge(df_comp, grupo_clusterFB1[['Distrito','DB P1']], how="outer", on=["Distrito"])
df_comp = pd.merge(df_comp, grupo_clusterFB2[['Distrito','DB P2']], how="outer", on=["Distrito"])
df_comp
```

```python
df_comp['Match']= df_comp['KM Total'] == df_comp['KM P1']
print("Variación de valores del Total al P1 en K-Means: ")
print(df_comp['Match'].value_counts()[False]/200*100, "%")

df_comp['Match']= df_comp['HAC Total'] == df_comp['HAC P1']
print("Variación de valores del Total al P1 en Jerárquico Aglomerativo: ")
print(df_comp['Match'].value_counts()[False]/200*100, "%")

df_comp['Match']= df_comp['DB Total'] == df_comp['DB P1']
print("Variación de valores del Total al P1 en DBSCAN: ")
print(df_comp['Match'].value_counts()[False]/200*100, "%")
```

```python
df_comp['Match']= df_comp['KM Total'] == df_comp['KM P2']
print("Variación de valores del Total al P2 en K-Means: ")
print(df_comp['Match'].value_counts()[False]/200*100, "%")

df_comp['Match']= df_comp['HAC Total'] == df_comp['HAC P2']
print("Variación de valores del Total al P2 en Jerárquico Aglomerativo: ")
print(df_comp['Match'].value_counts()[False]/200*100, "%")

df_comp['Match']= df_comp['DB Total'] == df_comp['DB P2']
print("Variación de valores del Total al P2 en DBSCAN: ")
print(df_comp['Match'].value_counts()[False]/200*100, "%")
```
