```python
import pandas as pd
```

```python
df = pd.read_csv('RawBased_2009to2013.csv', sep=',')
#df = df.drop(columns=['Unnamed: 0'])
dff = pd.read_csv('Scores1_FeatureBased.csv', sep=',')
dff = dff.drop(columns=['Unnamed: 0'])
```

```python
df1= pd.DataFrame()
lista1 = []
lista2 = []
#Definición de 1ra Columna
for i in list(range(4)):
    lista1.append('Jerárquico Aglomerativo')
for i in list(range(4)):
    lista1.append('K-means')
for i in list(range(4)):
    lista1.append('DBScan')   
    
#Definición de 2da Columna
for i in list(range(3)):
    lista2.append('Distancia Euclideana')
    lista2.append('Correlación Pearson')
    lista2.append('Correlación Spearman')
    lista2.append('Dynamic Time Warping')

#Agrupamiento de columnas para la tabla
df1['Algoritmo de agrupamiento'] = lista1
df1['Métrica de similaridad'] = lista2

lista3 = []

#Definición de 3ra Columna
for i in list(range(3)):
    lista3.append(df.iloc[0,i+1])
    lista3.append(df.iloc[1,i+1])
    lista3.append(df.iloc[2,i+1])
    lista3.append(df.iloc[3,i+1])
df1['Puntaje Silhouette RawBased'] = lista3

lista4 = []
#Definición de 3ra Columna
for i in list(range(3)):
    lista4.append(dff.iloc[0,i+1])
    lista4.append(dff.iloc[1,i+1])
    lista4.append(dff.iloc[2,i+1])
    lista4.append(dff.iloc[3,i+1])
df1['Puntaje Silhouette FBased'] = lista4
df1
```
